## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# your code here
def generate_uniform(bottom, ceiling, count):
    return stats.uniform.rvs(bottom, ceiling, count)
```

```python
def get_norm(mean, stdev, count):
    return stats.norm.rvs(mean,stdev,count)
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameters
mean_time = 10  # mean in minutes
lambda_param = 1 / mean_time

# Define the exponential distribution
dist = expon(scale=mean_time)

# Probability that time < 15 minutes
t = 15
prob = dist.cdf(t)
print(f"Probability that a customer spends less than {t} minutes in the bank: {prob:.4f}")

# Plot PDF and CDF
x = np.linspace(0, 40, 400)
pdf = dist.pdf(x)
cdf = dist.cdf(x)

plt.figure(figsize=(12, 5))

# PDF
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.title("Exponential Distribution - PDF")
plt.xlabel("Time (minutes)")
plt.ylabel("Probability Density")
plt.axvline(t, color='red', linestyle='--', label=f't = {t}')
plt.legend()

# CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='green')
plt.title("Exponential Distribution - CDF")
plt.xlabel("Time (minutes)")
plt.ylabel("Cumulative Probability")
plt.axvline(t, color='red', linestyle='--')
plt.axhline(prob, color='red', linestyle='--', label=f'P(T < {t}) = {prob:.4f}')
plt.legend()

plt.tight_layout()
plt.show()
```
